```python
import requests, numpy as np, pandas as pd
from datascience import *
```

```python
table = Table().read_table('districts.csv')

districts = []
for i in table.column(2):
    districts.append(i.replace(' ', '+'))
```

```python
def get_key(search):
    '''Scrapes webpage for a given search and returns the district ID'''
    r = requests.get(search)
    t = r.text
    return t[t.find('district_detail.asp'):t.find('"', t.find('district_detail.asp'))]
```

```python
search_template = 'https://nces.ed.gov/ccd/districtsearch/district_list.asp?Search=1&details=1&InstName=**DIST_NAME**'
final_template = 'https://nces.ed.gov/ccd/districtsearch/**KEY**&details='

keys = []
for i in districts:
    print(i)
    keys.append(get_key(search_template.replace('**DIST_NAME**', i)))
    
urls = []
for i in keys:
    urls.append(final_template.replace('**KEY**', i).replace('&details=1', ''))
    print(i)
        
```

```python
districts = Table().with_columns('Name', districts, 'URL', urls)
districts.to_df().to_csv('SchoolDistrictURLs.csv', index = False)
```
